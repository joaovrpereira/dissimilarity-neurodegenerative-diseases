```python
import monai
import monai.transforms as mt
import matplotlib.pyplot as plt
import ignite
import numpy as np
import torch
import warnings

warnings.filterwarnings("ignore") 

monai.config.print_config()
```

```python
#Monai supõe que os inputs são arrays numpy, pytorch tersors ou objetos tipo tensor.

KEYS = ("image", "label")

xform = mt.Compose(
    [
        mt.LoadImageD(KEYS),
        mt.EnsureChannelFirstD("image"),
        mt.EnsureChannelFirstD("label", channel_dim="no_channel"),
        mt.OrientationD(KEYS, axcodes="RAS"),
        mt.SpacingD(KEYS, pixdim=(1.0, 1.0, 1.0), mode=("bilinear", "nearest")),
        mt.ScaleIntensityD(keys="image"),
        mt.ResizeD(KEYS, (64, 64, 32), mode=("trilinear", "nearest")),
        mt.RandAffineD(
            KEYS,
            spatial_size=(-1, -1, -1),
            rotate_range=(0, 0, np.pi / 2),
            scale_range=(0.1, 0.1),
            mode=("bilinear", "nearest"),
            prob=1.0,
        ),
    ]
)

image_path1= "/mnt/hd/oasis-dataset/OAS31474_MR_d0150/anat2/sub-OAS31474_sess-d0150_T1w.nii.gz"
image_path2= "/mnt/hd/oasis-dataset/OAS30001_MR_d0757/anat2/sub-OAS30001_ses-d0757_run-01_T1w.nii.gz"

dict_loader = mt.LoadImageD(keys=("image", "label")) #retorna um MetaTensor 

data_list = [
    {"image":image_path1,"label":image_path1},
    {"image":image_path2,"label":image_path2}
]

dados_carregados = [dict_loader(dicionario) for dicionario in data_list]

print(dados_carregados[0]['image'].shape)
        
print(dados_carregados[1]['image'].shape)
```

```python
# Pré-Processamento
import os
import re
import nibabel as nib

    
data_dicts = []
spacings = []

# Padronização do tamanho dos voxels


def analyzeMetaData():
    
    t1FilePattern = r".*T1.*\.nii\.gz$"
    pattern = re.compile(t1FilePattern)
    print(pattern)
    directory = "/mnt/hd/oasis-dataset/"

    
    for dirpath, dirnames, filenames in os.walk(directory):
        for filename in filenames:
            if pattern.search(filename):
                t1_filepath = os.path.join(dirpath, filename)
                
                try:
                    img_obj = nib.load(t1_filepath)
                except Exception as e:
                    print(f"Não foi possível carregar o arquivo: {t1_filepath}\nErro: {e}")
                    continue 
                
                sample_dict = {
                    "image": t1_filepath,
                }

                data_dicts.append(sample_dict)
                voxel_spacing = img_obj.header.get_zooms()[:3] # Pega apenas os 3 eixos espaciais
                spacings.append(voxel_spacing)
                
                break# Para cada pasta, percorremos a lista de arquivos nela contidos
    

analyzeMetaData()
```

```python
sum = [0.0,0.0,0.0]
for dim in spacings:
    sum[0] += dim[0]
    sum[1] += dim[1]
    sum[2] += dim[2]

numeroImagens = len(spacings)
voxelMedio = [s / numeroImagens for s in sum]
print(f"O Voxel médio em cada canal é de: {voxelMedio}")
```
