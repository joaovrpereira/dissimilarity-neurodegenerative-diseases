```python
import pandas as pd
import re

def simplificar_dx1(texto):
    if pd.isna(texto):
        return "Unknown"

    # Normaliza texto
    t = str(texto).strip().lower()

    # Mapeamento simplificado por padro textual
    if re.search(r"\bad\b|\bad dem\b|\bdementia alzheimer\b", t):
        return "Alzheimer's Disease"
    elif re.search(r"\bvascular\b", t):
        return "Vascular Dementia"
    elif re.search(r"\bfrontotemporal\b|\bftd\b", t):
        return "Frontotemporal Dementia"
    elif re.search(r"\bdlbd\b|\blewy\b", t):
        return "Lewy Body Dementia"
    elif re.search(r"\bno dementia\b|\bcognitively normal\b|\bnormal\b", t):
        return "Cognitively Normal"
    elif re.search(r"\buncertain\b|\bunc\b|\bpossible\b|\bincipient\b", t):
        return "Uncertain / Mild Impairment"
    else:
        return "Other / Unknown"

# Aplicar a funo  coluna dx1
df[3]["dx1_simplified"] = df[3]["dx1"].apply(simplificar_dx1)

# Verificar contagens
print(df[3]["dx1_simplified"].value_counts())

# Salvar arquivo limpo
df[3].to_csv("OASIS3_dx1_simplified.csv", index=False)

print("\n Diagnsticos simplificados e arquivo salvo como 'OASIS3_dx1_simplified.csv'")
```
